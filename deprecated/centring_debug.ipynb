```python
%load_ext autoreload
%autoreload 2
import netCDF4
import xarray as xr
import hmp
import os
from pathlib import Path

DATA_PATH = Path(os.getenv("DATA_PATH"))
```

```python
data = xr.load_dataset(DATA_PATH / "sat1/data_100hz.nc")
data = data.isel(participant=[0,1,2,3,4])
print(data)
```

```python
data.sel(participant='0001', epochs=4, channels='Fp1').data
```

```python
ori_coords = data.coords

hmp_data = (
    data.stack(trial=["participant", "epochs", "channels"])
    .groupby("trial", squeeze=False)
    .map(hmp.utils._center)
    .unstack()
)
hmp_data.assign_coords(ori_coords)
```

```python
hmp_data.sel(participant="0001", epochs=4, channels="Fp1").to_numpy()
```
